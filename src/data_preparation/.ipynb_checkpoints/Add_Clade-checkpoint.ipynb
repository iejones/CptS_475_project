We want to find and add the nextstrain clade for every sequences. Nextstrain uses the nucleotide sequence, not the amino acid sequence we have. We will use the protein accession value in our data to search ncbi and retrieve the nucleotide accessions and sequences. Then we will will use the nextclade tool to find the clade for those sequences. We will join the nucleotide accession, sequence, and clade category to our origonal data table.

```python
# set up
import pandas as pd
from glob import glob
import os
project_folder = ".." + os.sep + ".." + os.sep
```

```python
# load our origonal data
origonal_dfs = []
for i in range(1, 12):
    path = os.path.join(project_folder, 'data', 'origonal', 'USA_T' + str(i) + '_*.csv')
    origonal_dfs.append(pd.read_csv(glob(path)[0]))
origonal_dfs[0].head(1)
```

First, we save a list of protein accessions for each time period into a file. We will use this to query the ncbi database.

```python
# make protien accession list files

accession_path = os.path.join(project_folder,"data","stage","accession")

i = 0
for df in origonal_dfs:
    i = i + 1
    df["accession"].to_csv(path_or_buf= os.path.join(accession_path, "accession_T" + str(i) + ".csv"), index = False, header = False)

#split T5 and T8 in half so small enough for ncbi site
for i in [5,8]:
    with open(os.path.join(accession_path, "accession_T" + str(i) + ".csv")) as f:
        lines = f.readlines()
    split_index = len(lines) // 2 
    with open(os.path.join(accession_path, "accession_T" + str(i) + "_part1.csv"), "w") as f:
        f.writelines(lines[:split_index])
    with open(os.path.join(accession_path, "accession_T" + str(i) + "_part2.csv"), "w") as f:
        f.writelines(lines[split_index:])
```

Now, open [ncbi virus](https://www.ncbi.nlm.nih.gov/labs/virus/vssi/#/virus?SeqType_s=Protein&VirusLineage_ss=Sev).

*   Save (A) as "Ti_protien_acc.txt" for timespan i, (B) as "Ti_nucleotide_acc.txt", (C) as "sequences_Ti.fasta"

Now, open [nextclade](https://clades.nextstrain.org/results).
For each timespan, upload the associated fasta file we just created. Download the results as csv to the `data\stage\nextclade_downloads` folder named as nextclade_Ti.csv

Now we read in all those files we just downloaded (and saved in the data/stage/ncbi_downloads folder) and merge them with our origonal data for each timespan.

```python
# helper functions for reading in our files and attaching

def convert_fasta(path): 
    # convert fasta file to pandas dataframe
    with open(path, "r") as f:
        lines = f.readlines()
    d = {'n_acc':[],
          'seq':[]
        }
    cur_seq = ""
    for l in lines:
        if l[0] == ">":
            if cur_seq:
                d['seq'].append(cur_seq)
                cur_seq = ""
            d['n_acc'].append(l.split(".")[0][1:])
        else:
            cur_seq = cur_seq + l.strip()
    # add last seq
    d['seq'].append(cur_seq)

    df = pd.DataFrame(d)
    return df

def get_clade_lookup(file):
    # get nucleotide seqeunce and claue name dataframe from nextstrain csv
    clade_lookup_df =  pd.read_csv(file, sep=";")[["seqName", "clade"]]
    clade_lookup_df['n_acc'] = clade_lookup_df['seqName'].transform(lambda x: x.split('.')[0])
    return clade_lookup_df.drop("seqName", axis=1)

def attach_n_seq(origonal_df, accession_df, necleotide_df, clade_df):
    # attach the nucleotide sequence, accession, and clade category for each row, matching on the accessions.
    df = origonal_df.merge(accession_df, left_on = "accession", right_on="p_acc").drop("p_acc", axis = 1)
    df = df.merge(necleotide_df)
    df = df.merge(clade_df, left_on = "n_acc", right_on="n_acc")
    df = df.rename(columns={"sequence": "p_sequence", "seq": "n_sequence", "accession": "p_accession", "n_acc": "n_accession"}, errors="raise")
    return df
```

```python
final_dfs = []
stage_folder = os.path.join(project_folder,"data","stage")
for i in range(1, 12): 
    protien_acc = pd.read_csv(os.path.join(stage_folder, 'ncbi_downloads', 'T' + str(i) +'_protien_acc.txt'), names=["p_acc"])
    nucleotide_acc = pd.read_csv(os.path.join(stage_folder, 'ncbi_downloads', "T" + str(i) +"_nucleotide_acc.txt"), names=["n_acc"])
    pairs_df = nucleotide_acc.join(protien_acc)
    nucleotide_df = convert_fasta(os.path.join(stage_folder, 'ncbi_downloads', 'sequences_T' +str(i) + '.fasta'))
    clade_df = get_clade_lookup(os.path.join(stage_folder, "nextclade_downloads", "nextclade_T" + str(i) + ".csv"))
    final_dfs.append(attach_n_seq(origonal_dfs[i - 1], pairs_df, nucleotide_df, clade_df))

final_dfs[0].head(5)
```
