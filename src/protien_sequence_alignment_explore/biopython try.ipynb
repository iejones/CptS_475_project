```python
seqA = alignments[0].seqA
seqB = alignments[0].seqB
length = len(seqA)
l = []
for i in range(length):
    if seqA[i] != seqB[i]:
        if seqA[i] == '-':
            l.append((None,  i, seqB[i]))
        elif seqB[i] == '-':
            l.append((seqA[i], i, None))
        else:
            l.append((seqA[i] , i, seqB[i]))
print(l)
final_l = []
skip = False 
offset = 0
for i in range(len(l) - 1):
    if l[i][1] == l[i+1][1] - 1:
        if l[i][2] is None and l[i+1][0] is None:
            print(l[i], l[i+1])
            final_l.append((l[i][0], l[i][1] - offset, l[i+1][2]))
            offset = offset + 1
            skip = True
            continue
        elif l[i][0] is None and l[i+1][2] is None:
            print("potential mutation? depends on align but I dont think it would do b first")
        else:
            final_l.append((l[i][0], l[i][1] - offset, l[i][2]))
    elif not skip:
         final_l.append((l[i][0], l[i][1] - offset, l[i][2]))
    skip = False
print(final_l)
print(length - offset)
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from itertools import chain
# Create figure and axes
fig, ax = plt.subplots()
fig.set_size_inches(18.5, 2.25)
ax.set_title("QKS89879 difference from QXQ22044", fontsize = 'x-large',  fontweight='bold')
ax.axes.get_xaxis().set_visible(False)
ax.axes.get_yaxis().set_visible(False)
ax.set_xlim([-15,1500])
ax.set_ylim([0.5,1.5])

y_height = 1
ax.hlines(y=y_height, xmin=0, xmax=length - offset, linewidth=1, color='k')

up_factor = 0.03
# Create rectangle patches for subunits
rects = []
rects.append(patches.Rectangle((14, y_height - 3 * up_factor), 671, up_factor*6, linewidth=0, edgecolor='r', facecolor='#ffbfba', label = "S1"))
rects.append(patches.Rectangle((686, y_height - 3 * up_factor), 587, up_factor*6, linewidth=0, edgecolor='b', facecolor='#c2d6ff', label = "S2"))

# Add patches to the Axes
for r in rects:
    ax.add_patch(r)



x_insert =[]
x_delete =[]
x_mutate =[]
up = 1
up_factor = 0.2
labels = []
for change in final_l:
    if change[0] is None:
        # insertion
        x_insert.append(change[1])
        labels.append({"text":str(change[1]) + change[2], 'xy' :(change[1], y_height + up *up_factor ), 'fontsize': 'x-large', 'verticalalignment':'center', 'horizontalalignment':'center'})
        up = - up
    elif change[2] is None:
        # deletion
        x_delete.append(change[1])
        labels.append({"text": change[0] + str(change[1]), 'xy' :(change[1], y_height + up *up_factor ), 'fontsize': 'x-large', 'verticalalignment':'center', 'horizontalalignment':'center'})
        up = - up
    else:
        # mutation
        x_mutate.append(change[1])
        labels.append({"text" :change[0] + str(change[1]) + change[2], 'xy' :(change[1], y_height + up *up_factor ), 'fontsize' : 'x-large', 'verticalalignment':'center', 'horizontalalignment':'center'})
        up = - up
        

ax.plot(x_insert ,[y_height for y in x_insert], marker='P', markersize=10, linewidth = 0, label = 'insertion')
ax.plot(x_delete ,[y_height for y in x_delete], marker='X', markersize = 10, linewidth = 0, label = 'deletion')
ax.plot(x_mutate ,[y_height for y in x_mutate], marker='o', markersize= 8,linewidth = 0, label = 'mutation')
ax.legend(fontsize = 'x-large', loc= 'center right')

for l in labels:
    ax.annotate(**l)

# remove frame
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)

plt.show()
fig.savefig('test.png', dpi=100, facecolor=fig.get_facecolor())
# missing the background behind the title on the saved image
```
